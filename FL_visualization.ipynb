```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataframes = ['/content/client1.csv', '/content/client2.csv', '/content/client3.csv','/content/client4.csv','/content/client5.csv',
              '/content/client6.csv','/content/client7.csv','/content/client8.csv','/content/client9.csv','/content/client10.csv',
              '/content/client11.csv','/content/client12.csv','/content/client13.csv','/content/client14.csv','/content/client15.csv']
x = 'Personalisation(%)'
y = 'F1score'
title = 'F1-score vs Personalization Percentage'

def F1_mean(x, y, title, dataframes):
    fig = plt.figure(figsize=(10, 5))
    mean_f1 = []  # List to store the mean F1-score for each client

    for i, df_path in enumerate(dataframes):
        # Load data from CSV into a dataframe
        df = pd.read_csv(df_path)

        # Calculate the mean F1-score for each client and handle missing values
        client_mean = df[y].mean()
        if not np.isnan(client_mean):
            mean_f1.append(client_mean)

        # Plot the F1-score for each client at different levels of personalization
        plt.plot(df[x], df[y], label=f'Client {i+1}')

    # Calculate the overall mean F1-score across all clients
    overall_mean_f1 = np.mean(mean_f1)

    # Plot the overall mean F1-score as a dashed black line
    plt.axhline(overall_mean_f1, color='black', linestyle='dashed', label='Overall Mean F1-score')

    # Set the title and labels for the graph
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend()  # Add a legend to show the client and overall mean labels
    plt.show()

F1_mean(x, y, title, dataframes)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataframes = ['/content/client1.csv', '/content/client2.csv', '/content/client3.csv','/content/client4.csv','/content/client5.csv',
              '/content/client6.csv','/content/client7.csv','/content/client8.csv','/content/client9.csv','/content/client10.csv',
              '/content/client11.csv','/content/client12.csv','/content/client13.csv','/content/client14.csv','/content/client15.csv']
x = 'Personalisation(%)'
y = 'F1score'
title = 'F1-score vs Personalization Percentage'

def F1_mean(x, y, title, dataframes):
    fig, ax = plt.subplots(figsize=(10, 5))

    for i, df_path in enumerate(dataframes):
        # Load data from CSV into a dataframe
        df = pd.read_csv(df_path)

        # Calculate the mean F1-score for each client and handle missing values
        client_mean = df[y].mean()
        if not np.isnan(client_mean):
            # Plot the F1-score for each client at different levels of personalization
            color = plt.cm.tab20(i)  # Get a unique color for each client
            ax.plot(df[x], df[y], label=f'Client {i+1}', color=color)

    # Set the title and labels for the graph
    ax.set_title(title)
    ax.set_xlabel(x)
    ax.set_ylabel(y)

    # Move the legend outside of the plot area
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    plt.show()

F1_mean(x, y, title, dataframes)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def mean_performance_plot(y, dataframes):
    met = [pd.read_csv(d)[y] for d in dataframes]
    df = pd.DataFrame(met)
    mean = df.mean(axis=0)
    plt.plot([i for i in range(0, 110, 10)], mean)
    plt.xlabel('Percentage of Personalization')
    plt.ylabel('Mean ' + y)
    plt.title('Mean ' + y + ' vs. Percentage of Personalization')
    plt.show()

# Example usage
dataframes = ['/content/client1.csv', '/content/client2.csv', '/content/client3.csv', '/content/client4.csv', '/content/client5.csv',
              '/content/client6.csv', '/content/client7.csv', '/content/client8.csv', '/content/client9.csv', '/content/client10.csv',
              '/content/client11.csv', '/content/client12.csv', '/content/client13.csv', '/content/client14.csv', '/content/client15.csv']

mean_performance_plot('F1score', dataframes)
```
